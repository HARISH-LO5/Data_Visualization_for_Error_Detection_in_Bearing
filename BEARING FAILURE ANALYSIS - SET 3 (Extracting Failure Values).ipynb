```python
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
import numpy as np
import os

def skewness(x):
    n = len(x)
    a = np.sum((x - np.mean(x))**3)/n
    b = np.std(x,ddof =1)**3
    return a/b

def kurtosis(x):
    n = len(x)
    a = np.sum((x - np.mean(x))**4)/n
    b = np.std(x,ddof =1)**4
    return a/b - 3
    
#TEST SET 3 BEARING NO 1

set3bear1 = pd.DataFrame()
test_set = 3
bearing_no = 1
path = r'D:\Bearing project Dataset\3rd_test\4th_test\txt'
for filename in os.listdir(path):
    dataset = pd.read_csv(os.path.join(path,filename), sep = '\t', header = None)
    datarr = np.array(dataset.iloc[:,bearing_no-1])
    feature_matrix = np.zeros((1,9))
    temp = datarr
    feature_matrix[0,0] = np.max(temp)
    feature_matrix[0,1] = np.min(temp)
    feature_matrix[0,2] = np.mean(temp)
    feature_matrix[0,3] = np.std(temp, ddof =1)
    feature_matrix[0,4] = np.sqrt(np.mean(temp**2))
    feature_matrix[0,5] = skewness(temp)
    feature_matrix[0,6] = kurtosis(temp)
    feature_matrix[0,7] = feature_matrix[0,0]/feature_matrix[0,4]
    feature_matrix[0,8] = feature_matrix[0,4]/feature_matrix[0,2]
    df1= pd.DataFrame(feature_matrix)
    df1.index = [filename[:-3]]
    set3bear1 = pd.concat([set3bear1,df1])

set3bear1.columns = ['Max', 'Min', 'Mean', 'Standard Deviation', 'RMS', 'Skewness', 'Kurtosis', 'Crest Factor', 'Form Factor']
set3bear1.index = pd.to_datetime(set3bear1.index, format ='%Y.%m.%d.%H.%M')
set3bear1.to_csv('Bearing_values_bearingno{}_testset{}.csv'.format(bearing_no,test_set))
set3bear1
```
